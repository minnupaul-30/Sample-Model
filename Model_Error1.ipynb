```python
import numpy as np
import pandas as pd
import scipy
from subprocess import check_output
import matplotlib.pyplot as plt
import keras
import os, sys
from IPython.display import display
from IPython.display import Image as _Imgdis
from PIL import Image
import numpy as np
from time import time
from time import sleep
from numpy import array
from numpy import genfromtxt
```

```python
data = pd.read_csv('TrainingData.csv')
data
```

```python

nm_array = data.values
nm_array[0][2]
```

```python
labels = []
images = []
for i in nm_array:
    labels.append(i[1])
    images.append(i[2])  
```

```python
train_images = array(images)
train_labels = array(labels)
train_images
```

```python
from keras import models
from keras import layers
```

```python
network = models.Sequential()
network.add(layers.Dense(512, activation='relu', input_shape=(1,)))
network.add(layers.Dense(1, activation='softmax'))
```

```python
network.compile(optimizer='rmsprop',

loss='sparse_categorical_crossentropy',
metrics=['accuracy'])
```

```python
train_images = train_images.reshape((100))
network.fit(train_images, train_labels, epochs=5, batch_size=128)
```
