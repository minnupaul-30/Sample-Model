```python
folder = "./images"
data = []
onlyfiles = []
onlyfiles_names = [f for f in os.listdir(folder) if os.path.isfile(os.path.join(folder, f))]
for f in os.listdir(folder):
    file = "./images/"+f
    #print(file)
    img = Image.open(file)
    img.load()
    data.append(np.array(img, dtype = "int32"))
    
    
for i in onlyfiles_names:
    name_split = i.split('.')
    onlyfiles.append(name_split[0])
print(onlyfiles)
#for i in range(40, 42):
   # print(onlyfiles[i])
    #display(_Imgdis(filename=folder + "/" + onlyfiles[i], width=240, height=320))
```
